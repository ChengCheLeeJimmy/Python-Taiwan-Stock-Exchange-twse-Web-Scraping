```python
import twstock
stock = twstock.Stock('2330')
print(stock.price) # 印出前 31 天的收盤價
# 可參考證交所的網站資料來比對 https://www.twse.com.tw/zh/trading/historical/stock-day.html
```

```python
import pandas as pd
df = pd.DataFrame({'日期':stock.date,
                   '開盤價':stock.open,
                   '收盤價':stock.close,
                   '最高價':stock.high,
                   '最低價':stock.low})
df
```

```python
# fetch 方法可讀取指定期間的歷史資料
stock.fetch(2023,6)
```

```python
type(stock.fetch(2023,6))
```

```python
pd_fetch = pd.DataFrame(stock.fetch(2023,6))
pd_fetch
```

```python
# fetch_from 方法可讀取指定月份到現在的歷史資料
pd_fetch_from = pd.DataFrame(stock.fetch_from(2023,1))
pd_fetch_from
```

```python
# 即時股票資訊
# https://twstock.readthedocs.io/zh_TW/latest/reference/realtime.html
real_stock = twstock.realtime.get('2330')
real_stock
```
